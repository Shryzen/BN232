```python
import math

class GeometricObject:
    def __init__(self, color="white", filled=False):
        self._color = color
        self._filled = filled

    def getColor(self):
        return self._color

    def setColor(self, color):
        self._color = color

    def isFilled(self):
        return self._filled

    def setFilled(self, filled):
        self._filled = filled

    def __str__(self):
        return f"Color: {self._color}, Filled: {self._filled}"

class Triangle(GeometricObject):
    def __init__(self, side1=1.0, side2=1.0, side3=1.0, color="white", filled=False):
        super().__init__(color, filled)
        self._side1 = side1
        self._side2 = side2
        self._side3 = side3

    def getSide1(self):
        return self._side1

    def getSide2(self):
        return self._side2

    def getSide3(self):
        return self._side3

    def setSide1(self, side1):
        self._side1 = side1

    def setSide2(self, side2):
        self._side2 = side2

    def setSide3(self, side3):
        self._side3 = side3

    def getArea(self):
        s = (self._side1 + self._side2 + self._side3) / 2
        return math.sqrt(s * (s - self._side1) * (s - self._side2) * (s - self._side3))

    def getPerimeter(self):
        return self._side1 + self._side2 + self._side3

    def __str__(self):
        return f"Triangle: side1 = {self._side1} side2 = {self._side2} side3 = {self._side3}"

# Example program to use the Triangle class

def main():
    side1 = float(input("Enter side 1 of the triangle: "))
    side2 = float(input("Enter side 2 of the triangle: "))
    side3 = float(input("Enter side 3 of the triangle: "))
    color = input("Enter the color of the triangle: ")
    filled = bool(int(input("Enter 1 if the triangle is filled, 0 otherwise: ")))

    triangle = Triangle(side1, side2, side3, color, filled)

    print(f"Area: {triangle.getArea()}")
    print(f"Perimeter: {triangle.getPerimeter()}")
    print(f"Color: {triangle.getColor()}")
    print(f"Filled: {triangle.isFilled()}")
    print(triangle)

if __name__ == "__main__":
    main()
```

```python
class Person:
    def __init__(self, name, address, phone, email):
        self.name = name
        self.address = address
        self.phone = phone
        self.email = email

    def __str__(self):
        return f"Person: {self.name}"

class Student(Person):
    def __init__(self, name, address, phone, email, class_status):
        super().__init__(name, address, phone, email)
        self.class_status = class_status

    def __str__(self):
        return f"Student: {self.name}, Status: {self.class_status}"

class Employee(Person):
    def __init__(self, name, address, phone, email, office, salary):
        super().__init__(name, address, phone, email)
        self.office = office
        self.salary = salary

    def __str__(self):
        return f"Employee: {self.name}, Office: {self.office}, Salary: {self.salary}"

class Faculty(Employee):
    def __init__(self, name, address, phone, email, office, salary, rank):
        super().__init__(name, address, phone, email, office, salary)
        self.rank = rank

    def __str__(self):
        return f"Faculty: {self.name}, Rank: {self.rank}"

class Staff(Employee):
    def __init__(self, name, address, phone, email, office, salary, title):
        super().__init__(name, address, phone, email, office, salary)
        self.title = title

    def __str__(self):
        return f"Staff: {self.name}, Title: {self.title}"

# Test program
def main():
    person = Person("Alice", "123 Main St", "555-1234", "alice@example.com")
    student = Student("Bob", "456 Elm St", "555-5678", "bob@example.com", "Sophomore")
    employee = Employee("Charlie", "789 Oak St", "555-8765", "charlie@example.com", "A1", 50000)
    faculty = Faculty("David", "321 Pine St", "555-4321", "david@example.com", "B2", 70000, "Professor")
    staff = Staff("Eve", "654 Cedar St", "555-2345", "eve@example.com", "C3", 40000, "Manager")

    print(person)
    print(student)
    print(employee)
    print(faculty)
    print(staff)

if __name__ == "__main__":
    main()
```

```python
class Loan:
    def __init__(self, principal, interest_rate, duration):
        self.principal = principal
        self.interest_rate = interest_rate
        self.duration = duration

    def calculate_total_payment(self):
        return self.principal * (1 + self.interest_rate * self.duration)

    def __str__(self):
        return f"Loan: Principal = {self.principal}, Interest Rate = {self.interest_rate}, Duration = {self.duration}"

class PersonalLoan(Loan):
    def __init__(self, principal, interest_rate, duration, borrower_age, employment_status):
        super().__init__(principal, interest_rate, duration)
        self.borrower_age = borrower_age
        self.employment_status = employment_status

    def __str__(self):
        return f"PersonalLoan: Principal = {self.principal}, Borrower Age = {self.borrower_age}"

class BusinessLoan(Loan):
    def __init__(self, principal, interest_rate, duration, company_size, industry_type):
        super().__init__(principal, interest_rate, duration)
        self.company_size = company_size
        self.industry_type = industry_type

    def __str__(self):
        return f"BusinessLoan: Principal = {self.principal}, Company Size = {self.company_size}"

# Test program
def main():
    personal_loan = PersonalLoan(10000, 0.05, 5, 30, "Employed")
    business_loan = BusinessLoan(50000, 0.04, 10, "Small", "Technology")

    print(personal_loan)
    print(f"Total payment: {personal_loan.calculate_total_payment()}")
    print(business_loan)
    print(f"Total payment: {business_loan.calculate_total_payment()}")

if __name__ == "__main__":
    main()
```
